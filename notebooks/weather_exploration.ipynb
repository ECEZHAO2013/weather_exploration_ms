```python
## sum of all odd number from 1-100: 1+3+5+7 ...  The sum of the first odd numbers is always n^2

odd_sum = 0

for num in  range(1,101):
  if num % 2 != 0: ## "!=" means "is not equal to"; use the Modulo operator (%), which tells us the remainder of a division
    odd_sum = odd_sum + num

print(f"The total is: {odd_sum}")
```

```python
## Example CODE to pull some or historical weather with TOKEN

import requests
import pandas as pd
from google.colab import userdata
```

```python
# Load token from Colab Secret
TOKEN = userdata.get("NOAA_TOKEN")
```

```python
headers = {
    "token": TOKEN
}
# üìç Georgia example station (Atlanta Airport)
station_id = "GHCND:USW00013874"

url = "https://www.ncei.noaa.gov/cdo-web/api/v2/data"

params = {
    "datasetid": "GHCND",
    "stationid": station_id,
    "startdate": "2024-01-01",
    "enddate": "2024-01-10",
    "units": "standard",
    "limit": 1000
}

response = requests.get(url, headers=headers, params=params)
data = response.json()

# Convert to DataFrame
df = pd.DataFrame(data["results"])
```
